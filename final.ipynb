```python
import pandas as pd
from glob import glob 
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
filenames = glob('*.csv')
def clean_file(file):
    missing_values=['na','NA','Not','NOT','Not list','Not Listed','Not listed']
    df = pd.read_csv(file,na_values=missing_values) # make the missing values as nan 
    df = df.dropna(how='any',axis=0) # drop missing values
    df = df[['Year','Model','Price','Mileage','Extcolor','Milerange']] # change the order of columns               
    df.Price = df.Price.str.replace('$','').str.replace(',','') #remove ($ and ,)
    df['Price'] = df['Price'].astype(int) #convert to int
    df.Mileage = df.Mileage.str.replace(',','') # remove (,)
    df['Mileage'] = df['Mileage'].astype(int) # convert to int 
    df['Milerange'] = df['Milerange'].astype(float) # convert to float
    df['Year'] = df['Year'].astype(int)
    return df
lst = []
for file in filenames:
    clean = clean_file(file)
    lst.append(clean)
lst
```
